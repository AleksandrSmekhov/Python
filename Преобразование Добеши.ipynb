```python
import random

# Генерация начального сигнала
def generate_data(N, signal_type):
    
    if signal_type is 1:
        f = open('linear.txt', 'w')
        for val in range(N):
            f.write(str(val) + '\n')
        f.close()
    elif signal_type is 2:
        f = open('constant.txt', 'w')
        for _ in range(N):
            f.write('1\n')
        f.close()

    elif signal_type is 3:
        f = open('rand_int.txt', 'w')
        for _ in range(N):
            f.write(str(random.randint(1, 100000)) + '\n')
        f.close()

    elif signal_type is 4:
        f = open('rand_float.txt', 'w')
        for _ in range(N):
            f.write(str(round(random.random(), 4)) + '\n')
        f.close()
    
# Чтение сигнала
def read_data(signal_type):
    temp = []
    if signal_type is 1:
        with open("linear.txt") as f:
            for line in f:
                temp.append(int(line))
    elif signal_type is 2:        
        with open("constant.txt") as f:
            for line in f:
                temp.append(int(line))
    elif signal_type is 3:        
        with open("rand_int.txt") as f:
            for line in f:
                temp.append(int(line))
    elif signal_type is 4:        
        with open("rand_float.txt") as f:
            for line in f:
                temp.append(float(line))
    return temp

# Коеффициенты Добеши
def daub_coefficients(order):
    if order is 2:
        return [7.071067811865475E-01, 7.071067811865475E-01]
    elif order is 4:
        return [0.4829629131445341E+00, 0.8365163037378079E+00, 0.2241438680420133E+00, -0.1294095225512603E+00]
    elif order is 6:
        return [0.3326705529500826E+00, 0.8068915093110925E+00, 0.4598775021184915E+00,
               -0.1350110200102545E+00,-0.8544127388202666E-01, 0.3522629188570953E-01]
    elif order is 8:
        return [0.2303778133088965E+00, 0.7148465705529156E+00, 0.6308807679298589E+00, -0.2798376941685985E-01,
               -0.1870348117190930E+00, 0.3084138183556076E-01, 0.3288301166688519E-01, -0.1059740178506903E-01]
    elif order is 10:
        return [0.1601023979741929E+00, 0.6038292697971896E+00, 0.7243085284377729E+00, 0.1384281459013207E+00,
               -0.2422948870663820E+00,-0.3224486958463837E-01, 0.7757149384004571E-01,-0.6241490212798274E-02,
               -0.1258075199908199E-01, 0.3335725285473771E-02]
    elif order is 12:
        return [0.1115407433501094E+00, 0.4946238903984530E+00, 0.7511339080210953E+00, 0.3152503517091976E+00,
               -0.2262646939654398E+00,-0.1297668675672619E+00, 0.9750160558732304E-01, 0.2752286553030572E-01,
               -0.3158203931748602E-01, 0.5538422011614961E-03, 0.4777257510945510E-02,-0.1077301085308479E-02]
    elif order is 14:
        return [ 7.785205408500917E-02, 3.965393194819173E-01, 7.291320908462351E-01, 4.697822874051931E-01,
                -1.439060039285649E-01,-2.240361849938749E-01, 7.130921926683026E-02, 8.061260915108307E-02,
                -3.802993693501441E-02,-1.657454163066688E-02, 1.255099855609984E-02, 4.295779729213665E-04,
                -1.801640704047490E-03, 3.537137999745202E-04]
    elif order is 16:
        return [5.441584224310400E-02, 3.128715909142999E-01, 6.756307362972898E-01, 5.853546836542067E-01,
               -1.582910525634930E-02,-2.840155429615469E-01, 4.724845739132827E-04, 1.287474266204784E-01,
               -1.736930100180754E-02,-4.408825393079475E-02, 1.398102791739828E-02, 8.746094047405776E-03,
               -4.870352993451574E-03,-3.917403733769470E-04, 6.754494064505693E-04,-1.174767841247695E-04]
    elif order is 18:
        return [3.807794736387834E-02, 2.438346746125903E-01, 6.048231236901111E-01, 6.572880780513005E-01,
                1.331973858250075E-01,-2.932737832791749E-01,-9.684078322297646E-02, 1.485407493381063E-01,
                3.072568147933337E-02,-6.763282906132997E-02, 2.509471148314519E-04, 2.236166212367909E-02,
               -4.723204757751397E-03,-4.281503682463429E-03, 1.847646883056226E-03, 2.303857635231959E-04,
               -2.519631889427101E-04, 3.934732031627159E-05]
    elif order is 20:
        return [2.667005790055555E-02, 1.881768000776914E-01, 5.272011889317255E-01, 6.884590394536035E-01,
                2.811723436605774E-01,-2.498464243273153E-01,-1.959462743773770E-01, 1.273693403357932E-01,
                9.305736460357235E-02,-7.139414716639708E-02,-2.945753682187581E-02, 3.321267405934100E-02,
                3.606553566956169E-03,-1.073317548333057E-02, 1.395351747052901E-03, 1.992405295185056E-03,
               -6.858566949597116E-04,-1.164668551292854E-04, 9.358867032006959E-05,-1.326420289452124E-05]

# Определение индекса    
def correct_index(top, val, bottom = 0):
    wide = top + 1 - bottom
    if wide == 1:
        return bottom
    else:
        temp = (val - bottom) % wide
        if temp < 0:
            temp += abs(wide)
        return bottom + temp

# Прямое преобразование Добеши
"""
signal - входной сигнал
order - порядок преобразования (2, 4, 6, 8, .., 20)
last_len - длина сигнала с необходимым масштабом преобразования (2^p)
max_len - длина изначального сигнала (2^p)

Out:
res_signal - преобразованный сигнал
"""
def daub_transform(signal, order, last_len, max_len):
    c = daub_coefficients(order)
    
    cur_len = max_len
    coef_amount = order - 1
    
    res_signal = signal.copy()
    
    while last_len <= cur_len:
        z = [0] * cur_len
        
        i = 0
        for j in range(0, cur_len - 1, 2):
            for k in range(0, coef_amount, 2):
                j0 = correct_index(cur_len - 1, j + k)
                j1 = j0 + 1
                
                z[i] += c[k] * res_signal[j0] + c[k + 1] * res_signal[j1]
                z[i + cur_len // 2] += c[coef_amount - k] * res_signal[j0] - c[coef_amount - k - 1] * res_signal[j1]
            i += 1
        
        res_signal[0:cur_len] = z
            
        cur_len //= 2
        
    for i in range(0, len(res_signal)):
        res_signal[i] = round(res_signal[i], 4)

    return res_signal

# Обратное преобразование Добеши
"""
signal - входной преобразованный сигнал
order - порядок преобразования (2, 4, 6, 8, .., 20)
length - текущий масштаб преобразования, длина входного преобразованного сигнала (2^p)
max_len - длина изначального сигнала (2^p)

Out:
res_signal - раскодированный сигнал
"""
def inv_daub_transform(signal, order, length, max_len):
    c = daub_coefficients(order)
    
    cur_len = length // 2
    coef_amount = order - 1
    q = (coef_amount - 1) // 2
    
    res_signal = signal.copy()
    
    while cur_len < max_len:
        z = [0] * cur_len * 2
        
        j = 0
        for i in range(-q, cur_len - q):
            for k in range(0, coef_amount, 2):
                i0 = correct_index(cur_len - 1, i + k // 2)
                i1 = i0 + cur_len
                
                z[j] += c[coef_amount - k - 1] * res_signal[i0] + c[k + 1] * res_signal[i1]
                z[j + 1] += c[coef_amount - k] * res_signal[i0] - c[k] * res_signal[i1]
            j += 2
                   
        cur_len *= 2
        
        res_signal[0:cur_len] = z
    
    for i in range(0, len(res_signal)):
        res_signal[i] = round(res_signal[i], 4)
        
    return res_signal
```

```python
# Пакет NumPy

import numpy as np
        
# Генерация начального сигнала
def np_generate_data(N, signal_type):
    
    if signal_type is 1:
        temp = np.arange(N)
        np.save('linear.npy', temp)
        
    elif signal_type is 2:
        temp = np.ones(N)
        np.save('constant.npy', temp)

    elif signal_type is 3:
        temp = np.random.randint(1, 100000, N)
        np.save('rand_int.npy', temp)

    elif signal_type is 4:
        temp = np.around(np.random.uniform(0.0, 1.0, N), decimals = 4)
        np.save('rand_float.npy', temp)
        
# Чтение сигнала
def np_read_data(signal_type):
    temp = []
    if signal_type is 1:
        temp = np.load('linear.npy')
    elif signal_type is 2:        
        temp = np.load('constant.npy')
    elif signal_type is 3:        
        temp = np.load('rand_int.npy')
    elif signal_type is 4:        
        temp = np.load('rand_float.npy')
    return temp

def С_matrix_generate(c_list, cur_len, order):
    coef_amount = order - 1
    
    cl = cur_len - 1
    cl2 = cur_len // 2
    
    C = np.zeros((cur_len, cur_len))
    
    i1 = 0
    
    for j in range(0, cl, 2):
            for k in range(0, coef_amount, 2):
                j0 = correct_index(cl, j + k)
                i2 = i1 + cl2
                
                C[i1][j0] = c_list[k]
                C[i2][j0] = c_list[coef_amount - k]
                
                k1 = k + 1
                j1 = j0 + 1
                C[i1][j1] = c_list[k1]
                C[i2][j1] = -c_list[coef_amount - k1]
            i1 += 1
            
    return C
    

# Прямое преобразование Добеши, прямые действия
# last_len >= order, max_len <= 32768
def np_daub_transform(signal, order, last_len, max_len):
    c = daub_coefficients(order)
    
    cur_len = max_len
    
    res_signal = signal.copy()
    
    while last_len <= cur_len:
        z = res_signal[0:cur_len]
        
        C = С_matrix_generate(c, cur_len, order)
                
        z = np.dot(C, z)
        
        res_signal[0:cur_len] = z
            
        cur_len //= 2
        
    res_signal=np.around(res_signal, decimals = 4)

    return res_signal

# Обратное преобразование Добеши
# last_len >= order, max_len <= 32768
def np_inv_daub_transform(signal, order, length, max_len):
    c = daub_coefficients(order)
    
    cur_len = length
    
    res_signal = signal.copy()
    
    while cur_len <= max_len:
        z = res_signal[0:cur_len]
        
        C = С_matrix_generate(c, cur_len, order).transpose()
        
        z = np.dot(C, z)
        
        res_signal[0:cur_len] = z
        
        cur_len *= 2
        
    res_signal=np.around(res_signal, decimals = 4)
    
    return res_signal

# Прямое преобразование Добеши
def npc_daub_transform(signal, order, last_len, max_len):
    c = daub_coefficients(order)
    
    cur_len = max_len
    
    
    coef_amount = order - 1
    
    res_signal = signal.copy()
    
    while last_len <= cur_len:
        z = np.zeros(cur_len)
        
        a = order // 2
        cl = cur_len - 1
        cl2 = cur_len // 2
    
        C1 = np.zeros((cur_len, a))
        R1 = np.zeros((cur_len, a))
        C2 = np.zeros((cur_len, a))
        R2 = np.zeros((cur_len, a))
        
        i = 0
        for j in range(0, cl, 2):
            i1 = 0
            for k in range(0, coef_amount, 2):
                j0 = correct_index(cl, j + k)
                j1 = j0 + 1
                
                a = i + cl2
                C1[i][i1] = c[k]
                C1[a][i1] = c[coef_amount - k]
                R1[i][i1] = res_signal[j0]
                R1[a][i1] = res_signal[j0]
                C2[i][i1] = c[k + 1]
                C2[a][i1] = -c[coef_amount - k - 1]
                R2[i][i1] = res_signal[j1]
                R2[a][i1] = res_signal[j1]
                
                i1 += 1
            i += 1
        
        z = np.sum(C1 * R1 + C2 * R2, axis = 1)
        
        res_signal[0:cur_len] = z
            
        cur_len //= 2
        
    res_signal=np.around(res_signal, decimals = 4)

    return res_signal

# Обратное преобразование Добеши
def npc_inv_daub_transform(signal, order, length, max_len):
    c = daub_coefficients(order)
    
    cur_len = length // 2
    coef_amount = order - 1
    q = (coef_amount - 1) // 2
    
    res_signal = signal.copy()
    
    while cur_len < max_len:
        cl = cur_len * 2
        z = np.zeros(cl)
        
        a = order // 2
        C1 = np.zeros((cl, a))
        R1 = np.zeros((cl, a))
        C2 = np.zeros((cl, a))
        R2 = np.zeros((cl, a))
        
        j = 0
        for i in range(-q, cur_len - q):
            j1 = 0
            for k in range(0, coef_amount, 2):
                i0 = correct_index(cur_len - 1, i + k // 2)
                i1 = i0 + cur_len
                
                a = j + 1
                C1[j][j1] = c[coef_amount - k - 1]
                C1[a][j1] = c[coef_amount - k]
                R1[j][j1] = res_signal[i0]
                R1[a][j1] = res_signal[i0]
                C2[j][j1] = c[k + 1]
                C2[a][j1] = -c[k]
                R2[j][j1] = res_signal[i1]
                R2[a][j1] = res_signal[i1]
                
                j1 += 1
            j += 2
        
        z = np.sum(C1 * R1 + C2 * R2, axis = 1)

        cur_len *= 2
        res_signal[0:cur_len] = z
    
    res_signal=np.around(res_signal, decimals = 4)
        
    return res_signal
```

```python
%%px

def correct_index(top, val, bottom = 0):
    wide = top + 1 - bottom
    if wide == 1:
        return bottom
    else:
        temp = (val - bottom) % wide
        if temp < 0:
            temp += abs(wide)
        return bottom + temp

def parallel_cycle(indexes):
    z_ind = []
    i = 0
    z_ind.append(id * len(z))
    z_ind.append(id * len(z) + len(indexes))
    
    for j in indexes:
        for k in range(0, coef_amount, 2):
            j0 = correct_index(cur_len - 1, j + k)
            j1 = j0 + 1

            z[i] += c[k] * res_signal[j0] + c[k + 1] * res_signal[j1]
            z[i + len(indexes)] += c[coef_amount - k] * res_signal[j0] - c[coef_amount - k - 1] * res_signal[j1]
        
        i += 1
        
    z_ind.append((id + 1) * len(z))
    
    return z_ind
    
def parallel_cycle_inv(indexes):
    j = 0
    
    for i in indexes:
        for k in range(0, coef_amount, 2):
            i0 = correct_index(cur_len - 1, i + k // 2)
            i1 = i0 + cur_len
                
            z[j] += c[coef_amount - k - 1] * res_signal[i0] + c[k + 1] * res_signal[i1]
            z[j + 1] += c[coef_amount - k] * res_signal[i0] - c[k] * res_signal[i1]
        j += 2
```

```python
def daub_transform_parallel(signal, order, last_len, max_len):
    c = daub_coefficients(order)
    dview.push({'c':c})
    
    cur_len = max_len
    coef_amount = order - 1
    dview.push({'coef_amount':coef_amount})
    
    res_signal = signal.copy()
    
    dview.scatter('id', client.ids, flatten=True)
    
    while last_len <= cur_len:
        z = [0] * cur_len
        
        index_list = list(range(0, cur_len - 1, 2))
        
        dview.scatter('index_list', index_list)
        dview.push({'res_signal':res_signal})
        dview.scatter('z', z)
        dview.push({'cur_len':cur_len})
        
        %px inds = parallel_cycle(index_list)
        
        inds = dview.pull('inds')
        temp_z = dview.gather('z')
        
        for i in client.ids:
            a = inds[0][1] * i
            b = inds[0][1] * (i + 1)
            c = len(temp_z) // 2
            z[a:b] = temp_z[inds[i][0]:inds[i][1]]
            z[a + c:b + c] = temp_z[inds[i][1]:inds[i][2]]

        res_signal[0:cur_len] = z
            
        cur_len //= 2
        
    for i in range(0, len(res_signal)):
        res_signal[i] = round(res_signal[i], 4)

    return res_signal

def inv_daub_transform_parallel(signal, order, length, max_len):
    c = daub_coefficients(order)
    dview.push({'c':c})
    
    cur_len = length // 2
    coef_amount = order - 1
    q = (coef_amount - 1) // 2
    dview.push({'coef_amount':coef_amount})
    
    res_signal = signal.copy()
    
    dview.scatter('id', client.ids, flatten=True)
    
    while cur_len < max_len:
        z = [0] * cur_len * 2
        
        index_list = list(range(-q, cur_len - q))
        
        dview.scatter('index_list', index_list)
        dview.push({'res_signal':res_signal})
        dview.scatter('z', z)
        dview.push({'cur_len':cur_len})
        
        %px parallel_cycle_inv(index_list)
        
        z = dview.gather('z')
            
        cur_len *= 2
        
        res_signal[0:cur_len] = z
    
    for i in range(0, len(res_signal)):
        res_signal[i] = round(res_signal[i], 4)
        
    return res_signal
```

```python
import pandas as pd

def get_list(name):
    temp=[]

    with open(name) as f:
        for line in f:
            temp.append(float(line))
            
    return temp

gen_time_data_norm = np.array(get_list("Time/gen_data.txt")).reshape((len(lengths1), len(orders1)))
gen_time_data_np = np.array(get_list("Time/np_gen_data.txt")).reshape((len(lengths1), len(orders1)))
d_gen_time_data = gen_time_data_np - gen_time_data_norm

read_time_data_norm = np.array(get_list("Time/read_data.txt")).reshape((len(lengths1), len(orders1)))
read_time_data_np = np.array(get_list("Time/np_read_data.txt")).reshape((len(lengths1), len(orders1)))
d_read_time_data = read_time_data_np - read_time_data_norm

transform_time_data_norm = np.array(get_list("Time/transform_data.txt")).reshape((len(lengths1), len(orders1)))
transform_time_data_np = np.array(get_list("Time/np_transform_data.txt")).reshape((len(lengths1), len(orders1)))
transform_time_data_parallel = np.array(get_list("Time/parallel_transform_data.txt")).reshape((len(lengths1), len(orders1)))
d_transform_time_data_npn = transform_time_data_np - transform_time_data_norm
d_transform_time_data_npp = transform_time_data_parallel - transform_time_data_np
d_transform_time_data_pn = transform_time_data_parallel - transform_time_data_norm

inv_transform_time_data_norm = np.array(get_list("Time/inv_transform_data.txt")).reshape((len(lengths1), len(orders1)))
inv_transform_time_data_np = np.array(get_list("Time/np_inv_transform_data.txt")).reshape((len(lengths1), len(orders1)))
inv_transform_time_data_parallel = np.array(get_list("Time/parallel_inv_transform_data.txt")).reshape((len(lengths1), len(orders1)))
d_inv_transform_time_data_npn = inv_transform_time_data_np - inv_transform_time_data_norm
d_inv_transform_time_data_npp = inv_transform_time_data_parallel - inv_transform_time_data_np
d_inv_transform_time_data_pn = inv_transform_time_data_parallel - inv_transform_time_data_norm
```
